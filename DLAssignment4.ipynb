## Importing MNIST dataset from Scikit-learn datasets

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import Isomap
from sklearn.manifold import TSNE
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import fetch_mldata
mnist= fetch_mldata('MNIST original')
```

## Selecting random data from the complete data set for training

```python
np.random.seed(42)
count = 2000
index= np.random.permutation(60000)[:count]
X= mnist['data'][index]
y= mnist['target'][index]
```

## Reducing dimensions using isomap technique

```python
model = Isomap(n_components=2)
X_reduced = model.fit_transform(X)
```

## Plotting the isomap

```python
plt.figure(figsize=(10,7))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

# Using t-SNE technique to reduce the dimensions

```python
tsne = TSNE(n_components=2, random_state=42)
X_reduced = tsne.fit_transform(X)
```

## Plotting t-SNE dimension reduction

```python
plt.figure(figsize=(10,7))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

## Locally Linear Embedding technique to reduce dimensions

```python
X_lle_reduced = LocallyLinearEmbedding(n_components=2, random_state=42).fit_transform(X)
```

## Plotting the LLE dimension reduction

```python
plt.figure(figsize=(10,7))
plt.scatter(X_lle_reduced[:, 0], X_lle_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

## Using PCA to reduce the dimensions

```python
X_pca_reduced = PCA(n_components=2, random_state=42).fit_transform(X)
```

## Plotting PCA dimension reduction

```python
plt.figure(figsize=(10,7))
plt.scatter(X_pca_reduced[:, 0], X_pca_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

## Using Multidimensional Scalling (MDS) to reduce dimensions

```python
X_mds_reduced = MDS(n_components=2, random_state=42).fit_transform(X[:count])
```

## Plotting MDS dimension reduction

```python
plt.figure(figsize=(10,7))
plt.scatter(X_mds_reduced[:, 0], X_mds_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

## Using Linear Discriminant Analysis(LDA) for dimension reduction

```python
X_lda_reduced = LinearDiscriminantAnalysis(n_components=2).fit_transform(X, y)
```

## Plotting LDA dimension reduction

```python
plt.figure(figsize=(10,7))
plt.scatter(X_lda_reduced[:, 0], X_lda_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

## Different methods of dimension reduction have been applied to MNIST data and the visualization are presented in above plots. The t-SNE represents the data more clearly as compared to other techniques used, since most of the different numbers can be clearly identified.
